```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Reading the Dataset
```

```python
df = pd.read_csv(r'Visualization1.csv')
```

```python
df.head()
```

```python
# conversion of percentage to float for proper visulization
```

```python
def per_convert(marks_avg):
    new=[]
    for i in marks_avg:
        s = i[:-1]
        new.append(float(s))
    return new
```

```python

df['Marks Avg']=per_convert(df['Marks Avg'])
```

```python
df['Behavioral Avg']=per_convert(df['Behavioral Avg'])
```

```python
df['Attendence']=per_convert(df['Attendence'])
```

```python
df
```

```python
d = ['Attendence','Marks Avg','Behavioral Avg']
```

```python
df.head()
```

```python
# Take user input 
```

```python
i = int(input("Enter Roll no: "))

s = df.iloc[i-1]
```

```python
s.to_dict()
```

```python
p = {k:s[k] for k in s.to_dict() if k in d}
```

```python
print(p)
```

```python
# Below plot shows all parameters of perticularly asked student
# Box Plot
```

```python
plt.bar(range(len(p)), list(p.values()), align='center')
plt.xticks(range(len(p)), list(p.keys()))

plt.xlabel(str(df.iloc[i-1,1]))
plt.show()
```

```python
print([i for i in p.values()])
```

```python
# Pie Chart
```

```python
labels = [i for i in p]
sizes = [i for i in p.values()]
 
fig1, ax1 = plt.subplots()
explode = (0, 0, 0) 
ax1.pie(sizes, explode=explode, labels=labels,autopct='%1.1f%%',
        shadow=True)
plt.xlabel(str(df.iloc[i-1,1]))
plt.show()
```
