```python
# this is all the library we have imported to predict marks of all students

import numpy as np
from numpy import math
import pandas as pd
import matplotlib.pyplot as plt
# Sklearn library used to import linear regression
import sklearn
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
```

```python
# here we have imported our data set form computer data set name is Book2.csv
dataset=pd.read_csv(r'Book2.csv')
dataset
```

```python
# Here we use scatter plot to see the relation between marks and behavoiur, how they are dependent on each other

plt.scatter(dataset['Behavioral Avg'],dataset['Marks Avg'],alpha=0.5)
plt.xlabel('Behavioural')
plt.ylabel('Marks')
plt.show()
```

```python
# Here we use scatter plot to see the relation between marks and Attendence, how they are dependent on each other

plt.scatter(dataset['Attendence'],dataset['Marks Avg'],alpha=0.5)
plt.xlabel('Attendence')
plt.ylabel('Marks')
plt.show()
```

```python
# here we defined our dependent variable 'marks' which we want  to predict in future

dependent_variable='Marks Avg'
```

```python
# here we defined our independent variable 'attendence and behavoiur'  this code will print all the column 
independent_variable=dataset.columns.tolist()
independent_variable
```

```python
# this is code is to remove dependent variable 'marks' from the list so that only independent variable will be remained
independent_variable.remove(dependent_variable)
independent_variable
```

```python
independent_variable.remove('Name')
```

```python
independent_variable
```

```python
# here we have convert all the table rows and column values of the table into array of values
# stored them into x and y
x=dataset[independent_variable].values
y=dataset[dependent_variable].values
x
```

```python
# here we have split our data set into 2 parts training  data and testing data 
# test_size =0.3 means 30 percent of the values will be used for testing model
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
X_train
X_test
y_train
y_test
```

```python
# here using min  max scaler  and fit_transform we tried to transform our training data
scaler=MinMaxScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
X_train
X_test
```

```python
X_train[0:10]
```

```python
regressor=LinearRegression() # this is the regression model used to train the dataset
regressor.fit(X_train,y_train) # here the fit function is used to train the model 
 
```

```python
y_pred=regressor.predict(X_test) # predicting on the final data set
y_pred
```

```python
math.sqrt(mean_squared_error(y_test,y_pred)) # used for checking the mean square error to check the accuracy of the system
```

```python
regressor.score(X_test,y_test) # this is also used to check the accuracy of the system
```
