```python
# Below Code is the conversion of percentage to 0, 1 and 2 for better visualization
# 0 - 50 -> 0
# 50 - 75 -> 1
# 75 - 100 -> 2
```

```python
def converter01(num):
    new1=[]
    for i in num:
        if(i in range(0,50)):
            i=0
        elif (i in range(50,75)):
            i=1
        else:
            i=2
        new1.append(i)
    return new1
```

```python
df['Attendence']= converter01(df['Attendence'])
```

```python
# Below plot - Attendence(X-axis) with respect to Marks(Y-axis)
```

```python
sns.set_style("whitegrid")
sns.boxplot(x='Attendence',y='Marks Avg',data=df)
```

```python
# Above plot shows that students having attendence 50 to 70% scores upto 70 t0 74% and having attendence 75-100% scores upto 60 to 90%
```

```python
df['Behavioral Avg']= converter01(df['Behavioral Avg'])
```
